# DataPreparation

```python
#appling a function to get the names of car companies.
car_prices_df["car_company"]=car_prices_df.CarName.apply(lambda x:x.split(' ')[0])
car_prices_df
```

# Handling Dummies

```python
# Creating dummys for the variables in below list
lt=["fueltype","aspiration","doornumber","enginelocation"]
def mapper(x):
    return x.map({'gas':1,'diesel':0,'std':0,'turbo':1,'two':0,'four':1,'front':0,'rear':1})
car_prices_df[lt]=car_prices_df[lt].apply(mapper)
```

```python
car_prices_df
```

```python
#creating dummys for variables with more than two categories 
drivewheel=pd.get_dummies(car_prices_df.drivewheel,drop_first=True)
carbody=pd.get_dummies(car_prices_df.carbody,drop_first=True)
enginetype=pd.get_dummies(car_prices_df.enginetype,drop_first=True)
cylindernumber=pd.get_dummies(car_prices_df.cylindernumber,drop_first=True)
fuelsystem=pd.get_dummies(car_prices_df.fuelsystem,drop_first=True)
car_company=pd.get_dummies(car_prices_df.car_company,drop_first=True)
```

```python
#concatinating the dummies
car_prices_df=pd.concat([car_prices_df,drivewheel,enginetype,cylindernumber,fuelsystem,carbody,car_company],axis=1)
car_prices_df
```

```python
#drop the created dummies
lt=["drivewheel","carbody","enginetype","cylindernumber","fuelsystem","car_company"]
car_prices_df.drop(lt,axis=1,inplace=True)
```

```python
car_prices_df.drop(["CarName"],axis=1,inplace=True)
```

```python
car_prices_df.drop(["car_ID"],axis=1,inplace=True)
```

Splitting Data into train and test

```python
#import sklearn and split the data
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test=train_test_split(car_prices_df,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_train.shape
```

```python
#fitting the model
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
num_vars = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
plt.figure(figsize=(46,12))
sns.heatmap(df_train.corr(),annot=True,cmap="YlGnBu")
plt.show()
```

```python
#pop the target variable for prediction
y_train = df_train.pop('price')
X_train = df_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 20
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, 10) 
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
# Dropping highly correlated variables and insignificant variables
#X = X.drop('two', 1)
#X = X.drop('subaru', 1)l
#X = X.drop('l', 1)
#X = X.drop('ohcf', 1)
X = X_train_rfe.drop('twelve', 1)
```

```python
# Build a second fitted model
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
