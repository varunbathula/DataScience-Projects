# Implementing PCA

```python
#Importing PCA From SKlearn
from sklearn.decomposition import PCA
```

```python
pca=PCA(svd_solver="randomized",random_state=100)
```

```python
#Applying PCA to data set
pca.fit(Xtrain)
```

```python
pca.components_
```

```python
#Comparing PCA with the features
Pc_Components=pd.DataFrame({"PC1":pca.components_[0],"PC2":pca.components_[1],"Features":X.columns})
Pc_Components
```

```python
#Ploting Features againest PC1 and PC2
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(Pc_Components.PC1, Pc_Components.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(Pc_Components.Features):
    plt.annotate(txt, (Pc_Components.PC1[i],Pc_Components.PC2[i]))
plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (12,8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
PCA_Hopkins=df_pca_nonpca.drop(['country'],axis=1)
#PCA_Hopkins.columns=["PCA1","PCA2","PCA3","PCA4"]
```

```python
PCA_Hopkins
```

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
#Appling Hopkins Statistics to know weather the dataset is suitable for clustering
hopkins(PCA_Hopkins)
```

```python
# sum of squared distances for getting the elbow curve and finding the optimal k
from sklearn.cluster import KMeans
ssd = []
for num_clusters in list(range(1,20)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(PCA_Hopkins)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```
